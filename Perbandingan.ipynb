Barplot: Penjualan Kopi per Provins

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
url = "https://raw.githubusercontent.com/nadiennn/visualisasi-/refs/heads/main/Data_persebaran.csv"
df1 = pd.read_csv(url)

# Barplot
plt.figure(figsize=(12, 6))
sns.barplot(data=df1, x='Provinsi', y='Penjualan_Kopi (Kg)', palette='viridis')
plt.xticks(rotation=45)
plt.title("Penjualan Kopi per Provinsi (Januari 2025)")
plt.tight_layout()
plt.show()
```

Pie Chart: Distribusi Kopi Unggula

```python
# Hitung jumlah tiap jenis kopi unggulan
kopi_counts = df1['Kopi_Unggulan'].value_counts()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(kopi_counts, labels=kopi_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Distribusi Kopi Unggulan")
plt.show()
```

Lineplot: Tren Penjualan Kopi per Bulan

```python
# Gabungkan Bulan dan Tahun jadi satu kolom waktu
df1['Waktu'] = df1['Bulan'] + ' ' + df1['Tahun'].astype(str)

# Urutkan berdasarkan waktu (opsional: bisa pakai mapping bulan kalau perlu)
plt.figure(figsize=(14, 6))
sns.lineplot(data=df1, x='Waktu', y='Penjualan_Kopi (Kg)', hue='Provinsi', marker='o')
plt.xticks(rotation=45)
plt.title("Tren Penjualan Kopi per Bulan")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
url = "https://raw.githubusercontent.com/nadiennn/visualisasi-/refs/heads/main/Data_persebaran.csv"
df1 = pd.read_csv(url, sep=";")

# Ambil nama kolom pertama (Provinsi/Wilayah)
provinsi_col = df1.columns[0]

# Hitung jumlah data per provinsi
provinsi_counts = df1[provinsi_col].value_counts()

# --- Pie Chart ---
plt.figure(figsize=(8,8))
plt.pie(provinsi_counts, labels=provinsi_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)
plt.title("Distribusi Data per Provinsi")
plt.axis("equal")  # supaya lingkaran bulat
plt.show()
```
